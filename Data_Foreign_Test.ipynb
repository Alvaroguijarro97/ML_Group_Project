# Foreign Population in Germany by state

```python
# Upload the "12521-0021_flat.csv" file into the notebook and set it as a dataframe
from google.colab import files
uploaded = files.upload()
```

```python
import io
foreign_data = pd.read_csv(io.BytesIO(uploaded['12521-0021_flat.csv']), encoding="windows_1258")

foreign_data.info()

foreign_data
```

```python
# rename columns, cleaning data
  # renaming: "Statistik_Code"	"Statistik_Label"	"Zeit_Code"	"Zeit_Label"	"Zeit" are all the same for this table. They can be added as header/general info about the table, no need to keep them inside the table
  # renaming: 1_Auspraegung_Label = State of residence (in Germany); 2_Auspraegung_Label = Sex; 3_Auspraegung_Label = Citizenship (outside Germany); BEV027__Auslaender__Anzahl = Number of Foreigners in the state
foreign_data_renamed = foreign_data.rename(columns={"1_Auspraegung_Label": "State of residence (in Germany)", "2_Auspraegung_Label": "Sex", "3_Auspraegung_Label": "Citizenship (outside Germany)", "BEV027__Auslaender__Anzahl": "Number of Foreigners in the state"})
# filter data for rows and columns that matter in the given context
  # suggest filter: from here on, using only "insgesamt" for Sex, not distinguishing male/female
foreign_data_renamed_filtered = foreign_data_renamed.loc[foreign_data_renamed["Sex"] == "Insgesamt"]
  # filter for the renamed variables (excluding sex, as it is the same across all observations after the last step) 
foreign_data_renamed_filtered = foreign_data_renamed_filtered.filter(items=["State of residence (in Germany)", "Citizenship (outside Germany)", "Number of Foreigners in the state"])

# make observation unit "state of residence", widen "Citizenship"
foreign_data_renamed_filtered = pd.pivot(foreign_data_renamed_filtered, index = "State of residence (in Germany)", columns = "Citizenship (outside Germany)", values = "Number of Foreigners in the state")

# order columns by country_name, "insgesamt" as a final column 
foreign_data_renamed_filtered = foreign_data_renamed_filtered [["Belgien", "Dänemark", "Frankreich", "Niederlande", "Österreich", "Polen", "Schweiz", "Tschechien", "Insgesamt"]]


foreign_data_renamed_filtered
```

The "world cities" data set contains information of over 4 million unique cities and towns from every country in the world. The variables observed in the dataset represent the following: 


*   **city**: 	The name of the city/town as a Unicode string (e.g. Goiânia).
*   **city_ascii**: city as an ASCII string (e.g. Goiania). Left blank if ASCII representation is not possible
*   **lat**: The latitude of the city/town
*   **lng**: The longitude of the city/town.
*   **country**: 	The name of the city/town's country.
*   **iso2**:	The alpha-2 iso code of the country.
*   **iso3**: 	The alpha-3 iso code of the country.
*   **admin_name**: The name of the highest level administration region of the city town (e.g. a US state or Canadian province). Possibly blank.
*   **population**: An estimate of the city's urban population. Only available for some (prominent) cities. If the urban population is not available, the municipal population is used.
*   **id**: A 10-digit unique id generated by SimpleMaps.

More information regarding this dataset can be found in: https://simplemaps.com/data/world-cities

```python
# Create a dataset only containing EU countries and their cities
eu_countries = geography_data[geography_data.iso3.isin(Europe)]
eu_countries = eu_countries[(eu_countries.capital == 'admin') | (eu_countries.capital == 'primary')]
eu_countries.info()
eu_countries.sort_values(by='population', ascending = False).head(25)
```

```python
# Create dataset containing German cities, but only the capitals of the different german states. 
germany = geography_data[(geography_data.iso3 == 'DEU')] 
germany = germany[(germany.capital == 'admin') | (germany.city == 'Berlin')]
germany.info()
germany.sort_values(by=['city'], ascending = [True])
ger_capitals = germany[['city','lat','lng']].reset_index()
ger_capitals
```

```python
def haversine_vectorize(lon1, lat1, lon2, lat2):

    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    newlon = lon2 - lon1
    newlat = lat2 - lat1

    haver_formula = np.sin(newlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(newlon/2.0)**2

    dist = 2 * np.arcsin(np.sqrt(haver_formula ))
    km = 6367 * dist #6367 for distance in KM for miles use 3958
    return km
```

```python
#Berlin 1
#Create a Dataframe for each German State Capital City and calculate the distance between every European state Capital
Berlin_Distance = eu_countries[['city','country','iso3','admin_name','lat','lng']].copy().reset_index(drop='True')

#Get the coordinates of the German State Capital City
long_ori = germany[germany['city'] == 'Berlin'][['lng']] 
long_ori= long_ori['lng'].iat[0]
lat_ori = germany[germany['city'] == 'Berlin'][['lat']]
lat_ori= lat_ori['lat'].iat[0]
Berlin_Distance["lat_ori"] = lat_ori
Berlin_Distance["long_ori"] = long_ori
#Calculate the distance between cities
Berlin_Distance["Distance"] = haversine_vectorize(Berlin_Distance['lng'],Berlin_Distance['lat'],Berlin_Distance['long_ori'],Berlin_Distance['lat_ori'])
Berlin_Distance = Berlin_Distance.sort_values(by = 'Distance').reset_index(drop='True')
Dis_Ber = Berlin_Distance.groupby(['iso3']).mean(['distance']).sort_values(by = 'Distance').drop(columns = ['lat','lng','lat_ori','long_ori'])
Dis_Ber = Dis_Ber.rename(columns = {"Distance":"Berlin"}).pivot_table(columns = "iso3")
Dis_Ber
```

```python
# importing required modules
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
# Upload the "naming_convention.csv" from GitHub/ML_Group_Project/resources/naming_convention.csv 
from google.colab import files
uploaded = files.upload()
import io
df_naming_convention = pd.read_csv(io.BytesIO(uploaded['naming_convention.csv']), encoding="windows_1258")
```

```python
# Upload the "df_abortion_atlas.csv" from GitHub/ML_Group_Project/wrangled data/df_abortion_atlas.csv 
from google.colab import files
uploaded = files.upload()
import io
df_abortion_atlas = pd.read_csv(io.BytesIO(uploaded['df_abortion_atlas.csv']), encoding="windows_1258")
```

```python
# merge "naming_convention.csv" and "df_abortion_atlas.csv"
df_abortion_atlas_rename = df_abortion_atlas.rename(columns={"Unnamed: 0": "df_Abortion_Atlas"})
df_abortion_atlas_rename['df_Abortion_Atlas'] = df_abortion_atlas_rename['df_Abortion_Atlas'].replace({'Swizerland': 'Switzerland', 'UK (E, S & W)': 'UK (E S & W)'})

df_abortion_atlas_country_code = pd.merge(df_abortion_atlas_rename,
                             df_naming_convention,
                             on='df_Abortion_Atlas',
                             how='outer',
                             indicator=True)
df_abortion_atlas_country_code = df_abortion_atlas_country_code.set_index("iso3")
```

```python
# take data frame with data on geographical distance between German State's Capital Cities and countries in Europe
# Upload the "geographical_data" file from GitHub/ML_Group_Project/wrangled data/geographical_data

from google.colab import files
uploaded = files.upload()
import io
geographical_data = pd.read_csv(io.BytesIO(uploaded['geographical_data.csv']), encoding="windows_1258")
```

```python
#Rename Columns and Index to maintain consistency through out notebooks
df_geographical_data = geographical_data.rename(columns={"ï»¿": "German Cities"}).set_index("German Cities").filter(items = ['BEL','BGR','BIH','DNK','EST','FIN','FRA','GRC','IRL','ITA','HRV','LVA','LTU','LUX','MLT','NLD','NOR','AUT','POL','PRT','ROU','SWE','CHE','SRB','SVK','SVN','ESP','CZE','HUN','GBR','CYP'])
df_geographical_data = df_geographical_data.rename(index = {"Stuttgart":"Baden-Württemberg","Munich": "Bavaria","Berlin":"Berlin","Potsdam":"Brandenburg","Bremen":"Bremen","Hamburg":"Hamburg","Wiesbaden":"Hesse","Hannover":"Lower Saxony","Schwerin":"Mecklenburg Western Pomerania","DĂ¼sseldorf":"Northrhine-Westphalia","Mainz":"Rhineland Palatinate","SaarbrĂ¼cken":"Saarland","Dresden":"Sachsen","Magdeburg":"Sachsen-Anhalt","Kiel":"Schleswig-Holstein","Erfurt":"Thuringia"}).sort_index()
df_geographical_data
```

```python
# Unpivot df from wide to long format, each observation is one state/country combination (instead of each observation a state with several countries as variables per single observation) 
geographical_data_by_state_long_format = pd.melt(df_geographical_data.reset_index(), id_vars=["German Cities"],
        var_name = "Country", value_name = "Distance")

geographical_data_by_state_long_format
 
# create place holder values for new column "Country_Score", that will allow to insert for each observation the corresponding abortion score for the correct country 
geographical_data_by_state_long_format["Country_Score"] = geographical_data_by_state_long_format.loc[:,"Country"]

# sort this so that it lists for each state the respective Distance
geographical_data_by_state_sorted = geographical_data_by_state_long_format.sort_values(by=["German Cities", "Distance"])

# create an object that counts in a range of one number per country 
counter = np.arange(1, (len(df_geographical_data.columns))+1)
# create column for this counter called "number" in order to keep sorted order intact later on 
geographical_data_by_state_sorted ["number"] = np.resize(counter, geographical_data_by_state_sorted .shape[0])
# pivot ("long to wide") df so that each state once again is row (observation)
geographical_data_by_state_sorted  = geographical_data_by_state_sorted .pivot(index="German Cities", columns="number", values=["Distance", "Country", "Country_Score"])
geographical_data_by_state_sorted
```

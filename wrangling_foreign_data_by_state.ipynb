```python
# take data frame with data on foreign population in Germany by state
# Upload the "EU_12521-0021_flat.csv" file from GitHub/ML_Group_Project/data/EU_12521-0021_flat.csv

from google.colab import files
uploaded = files.upload()
import io
foreign_data = pd.read_csv(io.BytesIO(uploaded['EU_12521-0021_flat.csv']), encoding="windows_1258")
```

```python
# Creating Dataframe Census Data, taken from "Data_Population_Migrant_Foreign.ipynb"
census_df = pd.DataFrame(
    {
    "Baden-Württemberg": [10976000,	7154000,	3822000,	1885925],
    "Bavaria": [12920000,	9493000,	3427000,	2051505],
    "Berlin": [3621000,	2369000,	1251000,	848400],
    "Brandenburg": [2474000,	2246000,	228000,	149540],
    "Bremen": [674000,	418000,	256000,	140330],
    "Hamburg": [1828000,	1199000,	630000,	328375],
    "Hesse": [6201000,	3980000,	2221000,	1163800],
    "Mecklenburg Western Pomerania": [1581000,	1443000,	138000,	87410],
    "Lower Saxony": [7857000,	6066000,	1791000,	895490],
    "Northrhine-Westphalia": [17658000,	12137000,	5521000,	2815800],
    "Rhineland Palatinate": [4017000,	2905000,	1112000,	532230],
    "Saarland": [968000,	749000,	219000,	134350],
    "Sachsen": [3995000,	3622000,	373000,	244415],
    "Sachsen-Anhalt": [2144000,	1949000,	194000,	127670],
    "Schleswig-Holstein": [2857000,	2335000,	523000,	286270],
    "Thuringia": [2090000,	1912000,	177000,	126270]
    }
)

census_df = census_df.T

census_df = census_df.rename(
        columns={
        0: "Overall_Population",
        1: "Population without migrant background",
        2: "Population with migrant Background",
        3: "Foreign Population (migrant background, no German Passport)"
        }
)
```

```python
# rename columns, cleaning data
  # renaming: 1_Auspraegung_Label = State of residence (in Germany); 3_Auspraegung_Label = Citizenship (outside Germany); BEV027__Auslaender__Anzahl = Number of Foreigners in the state
foreign_data_renamed = foreign_data.rename(columns={"1_Auspraegung_Label": "State of residence (in Germany)", "3_Auspraegung_Label": "Citizenship (outside Germany)", "BEV027__Auslaender__Anzahl": "Number of Foreigners in the state"})
# filter data for rows and columns that matter in the given context
  # filter for the renamed variables

foreign_data_renamed_filtered = foreign_data_renamed.filter(items=["State of residence (in Germany)", "Citizenship (outside Germany)", "Number of Foreigners in the state"])

# make observation unit "state of residence", widen "Citizenship"
foreign_data_renamed_filtered = pd.pivot(foreign_data_renamed_filtered, index = "State of residence (in Germany)", columns = "Citizenship (outside Germany)", values = "Number of Foreigners in the state")

# order columns by country_name, excluding "insgesamt"  
foreign_data_renamed_filtered = foreign_data_renamed_filtered [["Belgien", "Bulgarien", "Bosnien und Herzegowina", "Dänemark", "Estland", "Finnland", "Frankreich", "Griechenland", "Irland", "Italien", "Kroatien", "Lettland", "Litauen", "Luxemburg", "Malta", "Niederlande", "Norwegen", "Österreich", "Polen", "Portugal", "Rumänien", "Schweden", "Schweiz", "Serbien", "Slowakei", "Slowenien", "Spanien", "Tschechien", "Ungarn", "Vereinigtes Königreich", "Zypern"]]

foreign_data_by_state = foreign_data_renamed_filtered.rename(
    columns={
        "Belgien": "BEL",
        "Bulgarien": "BGR",
        "Bosnien und Herzegowina": "BIH",
        "Dänemark": "DNK",
        "Estland": "EST",
        "Finnland": "FIN",
        "Frankreich": "FRA",
        "Griechenland": "GRC",
        "Irland": "IRL",
        "Italien": "ITA",
        "Kroatien": "HRV",
        "Lettland": "LVA",
        "Litauen": "LTU",
        "Luxemburg": "LUX",
        "Malta": "MLT",
        "Niederlande": "NLD",
        "Norwegen": "NOR",
        "Österreich": "AUT",
        "Polen": "POL",
        "Portugal": "PRT",
        "Rumänien": "ROU",
        "Schweden": "SWE",
        "Schweiz": "CHE",
        "Serbien": "SRB",
        "Slowakei": "SVK",
        "Slowenien": "SVN",
        "Spanien": "ESP",
        "Tschechien": "CZE",
        "Ungarn": "HUN",
        "Vereinigtes Königreich": "GBR",
        "Zypern": "CYP",
        },
    index={
        "Bayern": "Bavaria",
        "Hessen": "Hesse",
        "Mecklenburg-Vorpommern": "Mecklenburg Western Pomerania",	
        "Niedersachsen": "Lower Saxony",
        "Nordrhein-Westfalen": "Northrhine-Westphalia",
        "Rheinland-Pfalz": "Rhineland Palatinate",
        "Thüringen" : "Thuringia"
        })
# function dividing foreign_data_by_state by census_df["Overall_Population"
for x in foreign_data_by_state:
  foreign_data_by_state[x] = foreign_data_by_state[x]/census_df["Overall_Population"]

foreign_data_by_state
```

```python
# Unpivot df from wide to long format, each observation is one state/country combination (instead of each observation a state with several countries as variables per single observation) 
foreign_data_by_state_long_format = pd.melt(foreign_data_by_state.reset_index(), id_vars=["State of residence (in Germany)"],
        var_name = "Country", value_name = "Foreign_Population")

 # create place holder values for new column "Country_Score", that will allow to insert for each observation the corresponding abortion score for the correct country 
foreign_data_by_state_long_format["Country_Score"] = foreign_data_by_state_long_format.loc[:,"Country"]
# sort this so that it lists for each state the respective Foreign Population
foreign_data_by_state_sorted = foreign_data_by_state_long_format.sort_values(by=["State of residence (in Germany)", "Foreign_Population"])

# create an object that counts in a range of one number per country 
counter = np.arange(1, (len(foreign_data_by_state.columns))+1)
# create column for this counter called "number" in order to keep sorted order intact later on 
foreign_data_by_state_sorted["number"] = np.resize(counter, foreign_data_by_state_sorted.shape[0])
# pivot ("long to wide") df so that each state once again is row (observation)
foreign_data_by_state_sorted = foreign_data_by_state_sorted.pivot(index="State of residence (in Germany)", columns="number", values=["Foreign_Population", "Country", "Country_Score"])
foreign_data_by_state_sorted
```
